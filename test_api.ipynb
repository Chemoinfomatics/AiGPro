```python
import requests
```

```python
sample = [
    "C12NC(CC)=NCC[C@@H1]1CCCC2C[C@H1]=CN",
    "C1[C@@H1]2C(C=NCN[C@@H1]2CCC(F)F)C=CC=CNCC=C1CC",
    "C1/CC(C2=NCN[C@@H1]2C(F)F)C=CC=CCCC=C1CC",
    "C/CC1(C2=NCN[C@@H1]2C=CCF)C=CC=CNCC=C1CC",
    "C1=CC2C(C=C1)=c1ccccc1=C2C1C=C(c2ccccc2)c2cccc3c4ccccc4nc1c23",
]
```

```python
TARGET_API = "http://localhost:8097/predictaigpro"

response = requests.post(
    TARGET_API,
    json={"smile": sample, "global_scan": True, "aig_version": "PRO"},
    headers={"Content-Type": "application/json", "Authorization": "Bearer " + "imsecreat"},
)
response.json()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(response.json())
```

## Post Processing

```python
from aigpro.chem.descriptors import custom_add_endpoint


custom_add_endpoint
```

```python
from aigpro.core.metrics import ic50_to_pic50
from aigpro.core.metrics import pic50_to_ic50


ic50_to_pic50
```

```python
def compute_azos_metrics(row, colname="plog_endpoint"):  # noqa: D103
    _standard_value = pic50_to_ic50(row[colname], unit="nM")
    row["standard_value"] = _standard_value
    _value = custom_add_endpoint(row)
    return _value


df["AZOS_ago"] = df.apply(compute_azos_metrics, axis=1, args=("ago_pred",))
df["AZOS_anta"] = df.apply(compute_azos_metrics, axis=1, args=("anta_pred",))
```
